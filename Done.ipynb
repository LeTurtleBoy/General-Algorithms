```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('Datos.txt',sep="\t", header = None)
df.columns = ['fecha', 'peticiones', 'errores']
data_arr = pd.DataFrame()
```

```python
df['peticiones'] = df['peticiones'].astype('int64')
df['errores'] = df['errores'].astype('int64')
df['PorError'] = df['errores']*100 / df['peticiones'] 
print(data_arr.head())
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
fig, ax1 = plt.subplots(figsize=(22,12))
title ="Pruebas Terminal Bogota"
ax1.set_title(title,fontsize= 30)
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Pruebas', size = 20)

ax1.plot(df.fecha, df.peticiones, color='red',marker='x',linestyle= ':',markersize=10)
ax1.tick_params(axis='y')

ax1.plot(df.fecha, df.errores, color='green',marker='x',linestyle= ':',markersize=10)
ax1.tick_params(axis='y')

for tick in ax1.get_xticklabels():
    tick.set_rotation(90)

ax2 = ax1.twinx()
ax2.set_ylabel('porcentaje de error', size = 20)  # we already handled the x-label with ax1
ax2.plot(df.fecha,df.PorError, color='blue',marker='.', linestyle= ':',markersize=10)
ax2.tick_params(axis='y')
fig.tight_layout()
legend = ax1.legend(loc='upper left', shadow=True, fontsize='x-large')
legend2 = ax2.legend(loc='upper right', shadow=True, fontsize='x-large')
fig.savefig('Grafica.png', format='png', dpi=900)
plt.show()

```

```python
import matplotlib.pyplot as plt
from pandas.plotting import table

desc = df.describe()
plot = plt.subplot(111, frame_on=False)
plot.xaxis.set_visible(False) 
plot.yaxis.set_visible(False) 
table(plot, desc,loc='upper right')
plt.savefig('desc_plot.png', format='png', dpi=900)
```
